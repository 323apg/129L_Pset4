```python
import numpy as np
import matplotlib.pyplot as plt

def fixed_points_stability(r):
    x1 = 0  # First fixed point at x = 0
    x2 = (r - 1) / r  # Second fixed point at x = (r - 1) / r
    
    f_prime_x1 = r  # Derivative at x1
    f_prime_x2 = -r + 2  # Derivative at x2
    
    stability_x1 = "Stable" if abs(f_prime_x1) < 1 else "Unstable"
    stability_x2 = "Stable" if abs(f_prime_x2) < 1 else "Unstable"
    
    return (x1, stability_x1), (x2, stability_x2)

r_values = [1, 2, 3, 4]

results = []

for r in r_values:
    fixed_point_1, fixed_point_2 = fixed_points_stability(r)
    results.append((r, fixed_point_1, fixed_point_2))

# Display the results
for r, fp1, fp2 in results:
    print(f"For r = {r}:")
    print(f"  Fixed point 1: x = {fp1[0]}, Stability: {fp1[1]}")
    print(f"  Fixed point 2: x = {fp2[0]}, Stability: {fp2[1]}")
    print()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_map_iteration(r, x0=0.2, threshold=1e-6, max_iter=1000):
    x = x0
    iteration = 0
    x_values = [x] 

    while iteration < max_iter:
        x_next = r * x * (1 - x)
        x_values.append(x_next)

        # Check for convergence
        if abs(x_next - x) < threshold:
            break
        
        x = x_next
        iteration += 1
    
    return x_values, iteration

r_values = [2, 3, 3.5, 3.8, 4.0]
threshold = 1e-6  # Convergence threshold

# Iterate for each r and print the results
for r in r_values:
    x_values, iterations = logistic_map_iteration(r, 0.2, threshold)
    print(f"r = {r}:")
    print(f"  Converged after {iterations} iterations.")
    print(f"  Final value of x: {x_values[-1]}")
    print(f"  First 10 values: {x_values[:10]}...")  # Display first few values for clarity
    print()

plt.figure(figsize=(10, 8))

for r in r_values:
    x_values, _ = logistic_map_iteration(r, 0.2, threshold)
    plt.plot(x_values, label=f'r = {r}')

plt.title("Convergence of the Logistic Map for Different r Values")
plt.xlabel("Iterations")
plt.ylabel("x_n (Population)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def logistic_map_iteration(r, x0, threshold=1e-6, max_iter=1000):
    x = x0
    iteration = 0
    x_values = [x]  # List to store x values at each iteration

    while iteration < max_iter:
        x_next = r * x * (1 - x)
        x_values.append(x_next)

        # Check for convergence
        if abs(x_next - x) < threshold:
            break
        
        x = x_next
        iteration += 1
    
    return x_values, iteration

r_values = [2, 3, 3.5, 3.8, 4.0]
initial_conditions = [0.1, 0.3, 0.5]
threshold = 1e-6

for r in r_values:
    plt.figure(figsize=(10, 8))  # Create a new figure for each r
    
    for x0 in initial_conditions:
        x_values, _ = logistic_map_iteration(r, x0, threshold)
        plt.plot(x_values, label=f'x0 = {x0}')

    plt.title(f"Time Series of Logistic Map for r = {r}")
    plt.xlabel("Iterations")
    plt.ylabel("x_n (Population)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    
    # Show the plot for this r
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to iterate the logistic map
def logistic_map(r, x0, num_iter=1000, last_iter=100):
    x = x0
    x_values = []
    for _ in range(num_iter):
        x = r * x * (1 - x)
        if _ >= (num_iter - last_iter):
            x_values.append(x)
    return x_values

# Parameters
r_values = np.linspace(0, 4, 10000)  # r values from 0 to 4
x0 = 0.2  # Initial condition
num_iter = 1000  # Total number of iterations
last_iter = 100  # Number of iterations to plot after system reaches equilibrium

# Plotting the bifurcation diagram
plt.figure(figsize=(10, 8))
for r in r_values:
    x_values = logistic_map(r, x0, num_iter, last_iter)
    plt.plot([r] * len(x_values), x_values, ',k', alpha=0.25)  # Plot each point

plt.title("Bifurcation Diagram of the Logistic Map")
plt.xlabel("r (Control Parameter)")
plt.ylabel("x (Population)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def modified_logistic_map(r, gamma, x0=0.2, iterations=1000, last_n=100):
    x = x0
    trajectory = []
    for _ in range(iterations):
        x = r * x * (1 - x**gamma)
        if _ >= (iterations - last_n):
            trajectory.append(x)
    return trajectory

def find_bifurcation_point(gamma_values, r_start=2, r_end=5.0, r_steps=1000, x0=0.2, iterations=1000, last_n=100):
    bifurcation_points = np.full(len(gamma_values), np.nan)
    r_values = np.linspace(r_start, r_end, r_steps)
    
    for i, gamma in enumerate(gamma_values):
        prev_steady_states = None
        for r in r_values:
            # Get the steady state values for the current r and gamma
            steady_states = modified_logistic_map(r, gamma, x0, iterations, last_n)
            unique_states = np.unique(np.round(steady_states, decimals=4))  # Round to reduce numerical precision issues
            
            # Check if bifurcation occurs: unique states > 1 indicates oscillations
            if prev_steady_states is not None and len(unique_states) > 1:
                bifurcation_points[i] = r
                break
            prev_steady_states = unique_states
    
    return bifurcation_points

# Define  values and compute bifurcation points
gamma_values = np.linspace(0.5, 1.5, 200)
r_vals = find_bifurcation_point(gamma_values)

plt.figure(figsize=(8, 6))
plt.plot(gamma_values[:len(r_vals)], r_vals, color='b', marker='o')
plt.title("First Bifurcation Point vs Gamma")
plt.xlabel("Gamma ()")
plt.ylabel("First Bifurcation Point (r)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
